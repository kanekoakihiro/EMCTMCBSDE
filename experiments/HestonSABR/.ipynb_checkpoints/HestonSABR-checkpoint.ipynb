```julia
using Distributions, CSV, DataFrames, ProgressMeter
include("../../utils_slv.jl")
```

```julia
T = 1.0; r=0.01; S0 = 100.0; v0 = 0.4; α = 0.65; β = 0.7; ρ = 0.65; K = 100.0; η=0.9; θ₀=0.02;
slvm = HestonSABRSLV(S0::Float64, v0::Float64, ρ::Float64, r::Float64, β::Float64, η::Float64, α::Float64, θ₀::Float64);
Π = [1.0 ρ; ρ 1.0]; cholΠ=cholesky(Π).L

drift(x) = [slvm.ω(abs(x[1]),abs(x[2])), slvm.μ(abs(x[2]))]
diffusion(x) = Diagonal(
        [slvm.m(abs(x[2]))*slvm.Γ(abs(x[1])), slvm.σ(abs(x[2]))]
        )*cholΠ
R = 0.07;
driver(t, x, y, z) = (
    -r*max(y-z'*inv(diffusion(x)*cholΠ)*x, 0.0)
    -R*min(y-z'*inv(diffusion(x)*cholΠ)*x, 0.0)
    -z'*inv(diffusion(x)*cholΠ)*drift(x)
)
terminal(x) = max(x[1]-K,0)
bsde = BSDE(T, [S0, v0], drift, diffusion, driver, terminal);
```

```julia
struct SpreadGrid <: AbstractGrid
    grid
    N
    Nₗ
    Nᵣ
    function SpreadGrid(p1, p2)
        grid = vcat(p1.grid[1:end-1],p2.grid)
        new(grid, p1.N+p2.N-1, p1.N-1, p2.N-1)
    end
end
```

```julia
Nₓ = [100, 15]
Δₓ = bsde.X0./Nₓ
Nₜ = 50
g₁ = 1.0; g₂ = 1.0
```

```julia
domain = [[Δₓ[1], Inf], [Δₓ[2], Inf]];

grids = Array{AbstractGrid,1}(undef, 2)

p1 = TavellaRandallGrid(g₁, g₂, Δₓ[1], K1, bsde.X0[1], Int(Nₓ[1]÷2), Int(Nₓ[1]÷2))
p2 = TavellaRandallGrid(g₁, g₂, bsde.X0[1], K2, 2*bsde.X0[1]-Δₓ[1], Int(Nₓ[1]÷2), Int(Nₓ[1]÷2))
grids[1] = SpreadGrid(p1,p2)
ratio = (Δₓ[2] * Nₓ[2]) / (Δₓ[1] * Nₓ[1])
grids[2] = TavellaRandallGrid(g₁*ratio, g₂*ratio, Δₓ[2], bsde.X0[2], 2*bsde.X0[2]-Δₓ[2], Nₓ[2], Nₓ[2])
```

```julia
scheme = HochOst4(krylov=true, m=100); EXPINT=true
exc_start = time()
res = MethodOfLines(bsde, grids, Nₜ, scheme, EXPINT)
exc_stop = time()
print(exc_stop-exc_start)
sol = res[1]; s_grid = res[2]; Q = res[3];
```
